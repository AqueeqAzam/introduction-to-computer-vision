# `Read image`

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/numar_innayat.jfif')
re_img = cv2.resize(img, (900, 600))
cv2_imshow( re_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# `Read video`

```python
import cv2
from google.colab.patches import cv2_imshow

video = cv2.VideoCapture('video path')
while True:
  success, img = video.read()

  cv2_imshow(img)
  if cv2.waitKey(1) & 0xFF == ord('q'):
    break
video.release()
cv2.destroyAllWindows
```

# `webCam`

```python
import cv2
from google.colab.patches import cv2_imshow

# read webCam
web = cv2.VideoCapture(1)

# visualize webcam

while True:
  success, img = web.read()
  cv2_imshow(img)
  if cv2.waitKey(1) & 0xFF == ord('q'):
    break

web.release()
cv2.destroyAllWindows()
```

# `Colorspaces`

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/numar_innayat.jfif')
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# `Bluring`

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/numar_innayat.jfif')
k_size = 12
blr_img = cv2.blur(img, (k_size, k_size))
# guss_blr_img = cv2.GaussianBlur(img, (k_size, k_size))
cv2_imshow(blr_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# `Threshold`

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/numar_innayat.jfif')
_, ths_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) # Assign only the thresholded image to ths_img
# guss_blr_img = cv2.GaussianBlur(img, (k_size, k_size))
cv2_imshow(ths_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# `Edge detaction`

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/numar_innayat.jfif')
edge_img = cv2.Canny(img, 100, 100)
cv2_imshow(edge_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# `Drawing`

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/eq_girl2.JPG')
# line = cv2.line(img, (100, 100), (200, 150), color=(0, 255, 0))
rect = cv2.rectangle(img, (50, 5), (370, 450), color=(0, 255, 0))
# cv2_imshow(line)
cv2_imshow(rect)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
